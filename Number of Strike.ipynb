```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
length = 1000
height = 600

deviation_x = length / 2
deviation_y = length / 2
```

```python
def getRNN():
    rnn = np.random.randn()
    print(rnn)
    return rnn
```

```python
poin_x = []
poin_y = []
HIT = 0
N = 0
for i in range(100):
    x = getRNN() * deviation_x
    y = getRNN() * deviation_y
    poin_x.append(x)
    poin_y.append(y)

    if (x <= deviation_x and x >= -deviation_x) and (y <= deviation_y and y >= -deviation_y):
        HIT += 1
    N += 1

area_x = [-500,500,500,-500,-500]
area_y = [-300,-300,300,300,-300]

plt.plot(area_x,area_y,color = "red")
plt.scatter(poin_x,poin_y)
plt.show()

print(HIT, N)
print("Accuracy : ", (HIT / N) * 100, "%")
```
