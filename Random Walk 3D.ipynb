```python
import math
import random
import matplotlib.pyplot as plt
```

```python
step = 0

x = 0
y = 0
z = 0

dotx = []
doty = []
dotz = []

direction = []

F_L_R = [0.4, 0.3, 0.2,0.1]
F_L_R = [int(10 * i) for i in F_L_R]

while step <= 100:
    rn = random.randint(0, 9)

    if rn in range(F_L_R[0]):
        direction.append('F')
        y += 1
        dotx.append(x)
        doty.append(y)
        dotz.append(z)
    elif rn in range(F_L_R[0] , F_L_R[0] + F_L_R[1]):
        direction.append('L')
        x -= 1
        dotx.append(x)
        doty.append(y)
        dotz.append(z)
    elif rn in range(F_L_R[0] + F_L_R[1], F_L_R[0] + F_L_R[1],F_L_R[2]):
        direction.append('R')
        x += 1
        dotx.append(x)
        doty.append(y)
        dotz.append(z)
    else:
        direction.append('Up')
        z += 1
        dotx.append(x)
        doty.append(y)
        dotz.append(z)


    step += 1
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(dotx, doty, dotz, 'green')
plt.show()
direction
```
